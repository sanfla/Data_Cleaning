```python
import pandas as pd
```

```python
data = pd.read_csv('HSP.csv')
pd.set_option('display.max_columns', None)
data
```

```python
data.shape
```

```python
data_info = pd.DataFrame()

data_info['null'] = data.isnull().sum()
data_info['Tipe'] = data.dtypes
data_info['Unik'] = data.nunique()

data_info = data_info[data_info['null'] > 0]

data_info['Persentase Null'] = (data_info['null'] / len(data)) * 100
data_info.sort_values(by='Persentase Null', ascending=False, inplace=True)

data_info
```

```python
data=data.drop(['Id','MiscFeature','PoolQC','Fence','Alley', 'MasVnrType','YrSold', 'YearBuilt', 'YearRemodAdd', 'GarageYrBlt'], axis=1)
```

```python
Numerical_col=[]
Categorical_col=[]
for feature in data.columns:
    if data[feature].dtype!="object":
        Numerical_col.append(feature)
    else:
        Categorical_col.append(feature)

print(Categorical_col)
print(Numerical_col)
```

```python
for col in data[Categorical_col]:
    data[col].fillna('No', inplace=True)
```

```python
data['LotFrontage'].fillna(data['LotFrontage'].median(), inplace=True)
data['MasVnrArea'].fillna(0, inplace=True)
```

```python
data_info = pd.DataFrame()

data_info['null'] = data.isnull().sum()
data_info['Tipe'] = data.dtypes
data_info['Unik'] = data.nunique()

data_info = data_info[data_info['null'] > 0]

data_info
```

```python
for column in Numerical_col:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = (data[column] < lower_bound) | (data[column] > upper_bound)
    
    data = data[~outliers]
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

for col in data[Categorical_col]:
    data[col] = encode.fit_transform(data[col])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[['SalePrice']] = scaler.fit_transform(data[['SalePrice']])
```

```python
data = data.drop_duplicates()
```
