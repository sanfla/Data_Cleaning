```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('SHP.csv')
data
```

```python
data_info = pd.DataFrame(data.dtypes, columns=['Tipe'])
data_info['Unik'] = data.nunique()
data_info['null'] = data.isnull().sum()
data_info
```

```python
data = data.drop(['Unnamed: 0', 'PassengerId', 'Name'], axis=1)
```

```python
unique_values = {col: data[col].unique() for col in data.columns}
for col, values in unique_values.items():
    print(f"{col}: {values}")
```

```python
data = data.replace(['inf', '-inf', '-'], np.nan)
```

```python
def replace(value):
    if isinstance(value, str) and value.endswith(('P', 'S')):
        return value[-1]  
    return value

data['Cabin'] = data['Cabin'].apply(replace)
```

```python
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Age'] = data['Age'].astype(int)
```

```python
numerical_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

data[numerical_columns] = data[numerical_columns].apply(pd.to_numeric, errors='coerce')
data.replace([np.inf, -np.inf, '-'], np.nan, inplace=True)
data[numerical_columns] = data[numerical_columns].apply(lambda x: x.fillna(x.median()))
```

```python
data['Transported'] = data['Transported'].astype(int)
```

```python
Numerical_col=[]
Categorical_col=[]
for feature in data.columns:
    if data[feature].dtype!="object":
        Numerical_col.append(feature)
    else:
        Categorical_col.append(feature)

print(Categorical_col)
print(Numerical_col)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

for col in data[Categorical_col]:
    data[col] = encode.fit_transform(data[col])
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
distribution = data['Transported'].value_counts(normalize=True)
print(distribution)
```

```python
from imblearn.over_sampling import SMOTE

X = data.drop('Transported', axis=1)
y = data['Transported']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

data = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
distribution = data['Transported'].value_counts(normalize=True)
print(distribution)
```

```python
data = data.drop(['Age', 'VIP',	'RoomService', 'FoodCourt',	'ShoppingMall',	'Spa', 'VRDeck'], axis=1)
```
