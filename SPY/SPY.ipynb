```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('SPY.csv')
pd.set_option('display.max_columns', None)
data
```

```python
data_info = pd.DataFrame()

data_info['null'] = data.isnull().sum()
data_info['Tipe'] = data.dtypes
data_info['Unik'] = data.nunique()

data_info = data_info[data_info['null'] > 0]

data_info['Persentase Null'] = (data_info['null'] / len(data)) * 100
data_info.sort_values(by='Persentase Null', ascending=False, inplace=True)

data_info
```

```python
data = data.drop(['Unnamed: 0', 'TIDAL Popularity', 'Release Date', 'ISRC', 'Track', 'Album Name', 'Artist', 'Soundcloud Streams'], axis=1)
```

```python
data = data.replace(',', '', regex=True)
```

```python
data = data.replace(['inf', '-inf', '-'], np.nan)
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
data
```

```python
data['All Time Rank'].isnull().sum()
```

```python
data = data.apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python
Q1 = data['All Time Rank'].quantile(0.25)
Q3 = data['All Time Rank'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (data['All Time Rank'] < lower_bound) | (data['All Time Rank'] > upper_bound)
data = data[~outliers]
```

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = data.columns.difference(['All Time Rank'])

scaler = MinMaxScaler()
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])
```

```python
data = data.drop_duplicates()
```
