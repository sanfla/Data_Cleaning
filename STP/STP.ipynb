```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('STP.csv')
pd.set_option('display.max_columns', None)
data
```

```python
data = data.drop(['Unnamed: 0', 'index'], axis=1)
```

```python
unique_values = {col: data[col].unique() for col in data.columns}
for col, values in unique_values.items():
    print(f"{col}: {values}")
```

```python
data = data.replace(['inf', '-inf', '-'], np.nan)
```

```python
data['age'] = pd.to_numeric(data['age'], errors='coerce')
data['Medu'] = pd.to_numeric(data['Medu'], errors='coerce')
data['Fedu'] = pd.to_numeric(data['Fedu'], errors='coerce')
data['traveltime'] = pd.to_numeric(data['traveltime'], errors='coerce')
data['studytime'] = pd.to_numeric(data['studytime'], errors='coerce')
data['failures'] = pd.to_numeric(data['Fedu'], errors='coerce')
data['famrel'] = pd.to_numeric(data['famrel'], errors='coerce')
data['freetime'] = pd.to_numeric(data['freetime'], errors='coerce')
data['goout'] = pd.to_numeric(data['goout'], errors='coerce')
data['Dalc'] = pd.to_numeric(data['Dalc'], errors='coerce')
data['Walc'] = pd.to_numeric(data['Walc'], errors='coerce')
data['health'] = pd.to_numeric(data['health'], errors='coerce')
data['absences'] = pd.to_numeric(data['absences'], errors='coerce')
data['G1'] = pd.to_numeric(data['G1'], errors='coerce')
data['G2'] = pd.to_numeric(data['G2'], errors='coerce')
data['G3(Target)'] = pd.to_numeric(data['G3(Target)'], errors='coerce')
```

```python
Numerical_col=[]
Categorical_col=[]
for feature in data.columns:
    if data[feature].dtype!="object":
        Numerical_col.append(feature)
    else:
        Categorical_col.append(feature)

print(Categorical_col)
print(Numerical_col)
```

```python
data[Numerical_col] = data[Numerical_col].apply(lambda x: x.fillna(x.median()))
```

```python
Q1 = data['G3(Target)'].quantile(0.25)
Q3 = data['G3(Target)'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (data['G3(Target)'] < lower_bound) | (data['G3(Target)'] > upper_bound)
data = data[~outliers]
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for col in Categorical_col:
    data.loc[:, col] = label_encoder.fit_transform(data[col])
```

```python
churn_distribution = data['G3(Target)'].value_counts(normalize=True)
print(churn_distribution)
```

```python
data = data.drop_duplicates()
```
